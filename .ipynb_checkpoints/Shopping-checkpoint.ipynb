```python
from IPython.display import clear_output

# Prints the cart for the show option and when the user is finished
def show_cart(cart):
    clear_output()
    total = 0
    print('Cart:')
    print('='*50)
    
    # If the cart is not empty, make a receipt with the items in the cart
    # Else print an empty receipt
    if cart != {}:
        for item, item_list in cart.items():
            # Calculates the cart total and prints the itemized receipt
            total += item_list[0] * item_list[1]
            formatted_price = '{:.2f}'.format(item_list[1])
            print(f'    {item.title()}: ${formatted_price} X {item_list[0]}')
    
    formatted_total = '{:.2f}'.format(total)
    print('='*50)
    print(f'Total: ${formatted_total}')


def shopping():
    cart = {}
    
    # Loops until the user inputs 'quit'
    while True:
        choice = input("Would you like to add/remove/show/clear or quit?: ").lower()
        
        if choice == 'quit':
            break
        elif choice == 'add':
            item = input('What would you like to add?: ')
            # If the item doesn't exist in the cart
            if cart.get(item, 'DNE') == 'DNE':
                price = float(input(f'How much does {item} cost?: '))
                quantity = int(input(f'How many {item} would you like to add?: '))
                cart[item] = [quantity, price]
            # If it does exist in the cart
            else:
                cart[item][0] += 1
        elif choice == 'remove':
            item = input('What would item would you like to remove?: ')
            
            # If the item does not exist, then prompt the user, else delete the item
            if cart.get(item, 'DNE') == 'DNE':
                print(f'{item.title()} is not in the cart. Please try again.')
            else:
                del cart[item]
        
        # Show the items currently in the cart
        elif choice == 'show':
            show_cart(cart)
        # Clears all items from the cart
        elif choice == 'clear':
            cart.clear()
            show_cart(cart)
        # If the user did not enter a valid action
        else: 
            print('Not a valid action.')
    
    clear_output()
    show_cart(cart)
```

```python
shopping()
```
